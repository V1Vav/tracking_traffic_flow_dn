```python
from ultralytics import YOLO
import os
```

```python
CUR_DIR = os.getcwd()
DATA_NAME = "640-Hutech-Vehicle-1"
DATA_YAML = f"{CUR_DIR}\{DATA_NAME}\data.yaml"

# print(DATA_YAML)

assert os.path.exists(DATA_YAML), "Không tìm thấy data.yaml"
print("Dataset OK")
```

```python
MODEL_DIR = f"{CUR_DIR}\..\models"
model_name = "yolov8n.pt"

model = YOLO(f"{MODEL_DIR}\{model_name}")
print("Loaded pretrained model")
```

```python
from datetime import datetime

# config
EPOCHS = 200

# tạo tên theo date + epoch
date_str = datetime.now().strftime("%Y%m%d_%H%M")
run_name = f"traffic_{date_str}_e{EPOCHS}"

print("Run name:", run_name)
```

```python
model.train(
    data=DATA_YAML,
    epochs=EPOCHS,
    imgsz=640,
    batch=4,
    device=0,
    name=run_name,
    verbose=True
)
```
